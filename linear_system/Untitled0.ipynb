```python
x = [[3, 1, 2,], [2, 6, -1], [4, 0, -1]]
y = [5, 1, 3]

x0 = x[0]
x1 = x[1]
x2 = x[2]

y0 = y[0]
y1 = y[1]
y2 = y[2]

print("original matrix")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 1 x 1/3
tmp = 1 / x0[0]
x0 = [element * tmp for element in x0]
y0 = y0 * tmp

print("row 1 x 1/3")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 1 x - 2 + row 2
x0_tmp = [element * -x1[0] for element in x0]
y0_tmp = y0 * -x1[0]

for i in range(len(x[0])):
  x1[i] += x0_tmp[i]
y1 += y0_tmp

print("row 1 x - 2 + row 2")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 1 x -4 + row 3
x0_tmp = [element * -x2[0] for element in x0]
y0_tmp = y0 * -x2[0]

for i in range(len(x[0])):
  x2[i] += x0_tmp[i]
y2 += y0_tmp

print("row 1 x -4 + row 3")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 2 x 3/16
tmp = 1 / x1[1]
x1 = [element * tmp for element in x1]
y1 += tmp

print("row 2 x 3/16")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 2 x 4/3 + row 3
x1_tmp = [element * -x2[1] for element in x1]
y1_tmp = y1 * -x2[1]

for i in range(len(x[1])):
  x2[i] += x1_tmp[i]
y2 += y1_tmp

print("row 2 x 4/3 + row 3")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```

```python
# row 3 x (-1/4.2)
tmp = 1 / x2[2]
x2 = [element * tmp for element in x2]
y2 *= tmp

print("row 3 x (-1/4.2)")
print(x0, y0)
print(x1, y1)
print(x2, y2)
```
