```python
# Norm of Tensor
A = [   [[1, 0, 1, 0], [2, 1, 0, 1], [3, 4, 0, 2]],
        [[3, 1, 2, 0], [1, 0, 4, 2], [0, 1, 0, 2]]  ]

n = len(A[0])
p = len(A[0][0])
q = len(A)
print(n, p, q)

norm = 0
for i in range(n):
  for j in range(p):
    for k in range(q):
      norm += A[k][i][j]**2
norm = norm ** 0.5

print(norm)
```

```python
# Inner Product of Tensor
A = [   [[1, 0, 1, 0], [2, 1, 0, 1], [3, 4, 0, 2]],
        [[3, 1, 2, 0], [1, 0, 4, 2], [0, 1, 0, 2]]  ]
B = [   [[2, 1, 3, 1], [0, 2, 1, 3], [1, 0, 1, 1]],
        [[1, 2, 3, 1], [3, 1, 0, 3], [0, 2, 3, 1]]  ]

n = len(A[0])
p = len(A[0][0])
q = len(A)

inner_product = 0
for i in range(n):
  for j in range(p):
    for k in range(q):
      inner_product += A[k][i][j] * B[k][i][j]
  
print(inner_product)
```

```python
# Inner Product of Tensor by Numpy
import numpy as np

A = np.array([  [[1, 0, 1, 0], [2, 1, 0, 1], [3, 4, 0, 2]],
                [[3, 1, 2, 0], [1, 0, 4, 2], [0, 1, 0, 2]]  ])
B = np.array([  [[2, 1, 3, 1], [0, 2, 1, 3], [1, 0, 1, 1]],
                [[1, 2, 3, 1], [3, 1, 0, 3], [0, 2, 3, 1]]  ])

C = np.multiply(A, B)
print(C)

inner_product = np.sum(C)
print(inner_product)
```
