Scalar는 크기만으로 나타낼 수 있는 물리량

Scalar의 데이터셋은 하나의 구성 원소, 하나의 숫자

기본적인 변수를 이용한 연산

Vector는 Scalar의 집합

matrix를 구성하는 기본 단위

Vector은 크기와 방향 모두를 나타내는 개념

Scalar를 행 방향으로 나열 -> Row Vector

Scalar를 열 방향으로 나열 -> Column Vector

```python
#sum of vector
u = [1, 2, 3]
v = [4, 5, 6]

n = len(u)
w = []

for i in range(n):
  val = u[i] + v[i]
  w.append(val)

print(w)
```

```python
#mutiple of vector
def v_mul(u, v):
  w = []
  for i in range(len(u)):
    w.append(u[i] * v[i])
  return w

u = [1, 2, 3]
v = [4, 5, 6]
print(v_mul(u, v))
```

```python
#Using numpy
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])
w = u + v
print(w)
```
