Matrix Trace(행렬의 대각합)

정사각 행렬일때만 성립

A =

  a11 a12 a13

  a21 a22 a23

  a31 a32 a33

  tr(A) = a11 + a22+ a 33

```python
import numpy as np

A = np.array([[2, 7], [3, 4], [6,1]])
B = np.array([[1, 4], [4, -1], [2, 5]])

C = A + B
print(C)

C = A - B
print(C)

# 행렬의 원소 곱
C = np.multiply(A, B)
print(C)

# 행렬 곱
A = np.array([[2, 7], [3, 4], [5, 2]])
B = np.array([[3, -3, 5], [-1, 2, -1]])
C = np.matmul(A, B)
print(C)
```

Triangular Matrix(삼각 행렬)

Toepliz Matrix(토플리츠 행렬)

  - a b c d e
  - f a b c d
  - g f a b c
  - h g f a b
  - i h g f a

Ti,j = Ti+1,j+1 = ti-j

```python
from scipy.linalg import toeplitz

A = toeplitz([1, 0, -2, -4], [1, 3, 5, 7, 9])
print(A)
```

Bidiagonal Matrix(이중 대각 행렬)

대각 원소뿐만아니라 위(upper) 혹은 아래(lower) 원소가 0이 아닌 행렬

```python
A = np.array([[1, 2, 1, 3,], [5, 3, 4, 1], [2, 1, 7, 9], [2, 8, 1, 3]])
print(A)
```

```python
diag_ele = np.diag(A)
print(diag_ele)
```

```python
np.diag(diag_ele)
```

```python
u_diag_ele = np.diag(A, k=1)
print(u_diag_ele)
```

```python
np.diag(u_diag_ele, k=1)
```

```python
u_diag = np.diag(diag_ele) + np.diag(u_diag_ele, k=1)
print(u_diag)
```

Householder matrix(하우스홀더 행렬)

정사각 행렬이며 모든 열이 정규 직교(orthonomal)하는 행렬

H = I - 2 * vvT/vTv


```python
import numpy as np

v = np.array([1, 0, 2, 3])
n = len(v)
#vvT
outer_mat = np.outer(v, v)
print(outer_mat)
```

```python
#vTv
inner_val = np.inner(v, v)
print(inner_val)
```

```python
#단위 행렬 생성
I = np.identity(n)
print(I)
```

```python
H = I - (2/inner_val)*outer_mat
print(H)
```
