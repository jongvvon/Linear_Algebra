깊은 복사와 얕은 복사

파이썬은 모두 객체로 이루어짐
객체의 종류는 mutable 과 imtuable 로 이루어져있음
 - mutable : 변경 가능한 객체
  -> 리스트
 - immutable : 변경 불가능한 객체
  -> 숫자, 문자

```python
# 깊은 복사 예제
x = 3
y = 3
print(x)
print(y)
print(id(x))
print(id(y))
```

```python
y = 7
print(x)
print(y)
print(id(x))
print(id(y))
```

숫자형 자료의 경우 immutable한 객체기 때문에 y값 변경과 상관없이 x의 값은 유지되는 것을 볼 수 있음

```python
a = [1, 2, 3, 4]
id(a)
```

```python
id(a[0])
```

```python
a[0] = 5
print(a)
print(id(a))
print(id(a[0]))
```

리스트는 mutable 한 객체인 반면 리스트내의 원소들은 immutalble 한 것을 볼 수 있음

```python
b = a
print(b)
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3, 4, 5]
b = a[:]
print(id(a))
print(id(b))
```

```python
import copy
a = [[1, 3], [2, 4]]
b = copy.copy(a)
print(a)
print(b)
print(id(a))
print(id(b))
print(id(a[0]))
print(id(b[0]))
```

```python
import copy
a = [[1, 3], [2, 4]]
b = copy.deepcopy(a)
print(a)
print(b)
print(id(a))
print(id(b))
print(id(a[0]))
print(id(b[0]))
```
